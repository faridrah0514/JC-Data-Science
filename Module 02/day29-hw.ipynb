```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
xl_data=pd.read_csv('misc/latihan_ujian_datavis_jcds04/EXCL.JK.csv', parse_dates=['Date'])
smartfren_data=pd.read_csv('misc/latihan_ujian_datavis_jcds04/FREN.JK.csv', parse_dates=['Date'])
indosat_data=pd.read_csv('misc/latihan_ujian_datavis_jcds04/ISAT.JK.csv',parse_dates=['Date'])
telkom_data=pd.read_csv('misc/latihan_ujian_datavis_jcds04/TLKM.JK.csv', parse_dates=['Date'])

telkom_data.set_index('Date', inplace=True)
xl_data.set_index('Date', inplace=True)
smartfren_data.set_index('Date', inplace=True)
indosat_data.set_index('Date', inplace=True)

new=telkom_data['2019-03':'2019-06'].resample('W').sum().copy()
plt.figure(figsize=(15,6))
plt.plot(xl_data['2019-03':'2019-06'].index, xl_data['2019-03':'2019-06']['Close'], label='PT XL Axiata Tbk', color='g')
plt.plot(smartfren_data['2019-03':'2019-06'].index, smartfren_data['2019-03':'2019-06']['Close'], label='PT Smartfren Telecom Tbk')
plt.plot(indosat_data['2019-03':'2019-06'].index, indosat_data['2019-03':'2019-06']['Close'], label='PT Indosat Tbk', color='b')
plt.plot(telkom_data['2019-03':'2019-06'].index, telkom_data['2019-03':'2019-06']['Close'], label='PT Telekomunikasi Indonesia Tbk', color='red')
plt.title('Harga Historis Saham Provider Telco Indonesia', fontdict={'fontsize':25, 'fontweight':'bold'}, loc='center',in_layout=False, va='top', position=(0.5, 1.15))
plt.legend(loc='upper center',frameon=False, bbox_to_anchor=(0, 1.02, 1, 0.06), mode='expand', ncol=4)

plt.xlabel('Tanggal', fontsize=15)
plt.ylabel('Rupiah (IDR)', fontsize=15)
plt.xticks(new.index,rotation=30)
plt.grid()
```
