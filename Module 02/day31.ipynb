# matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as im
```

```python
df = pd.read_csv('misc/dataSales.csv')
df=df.pivot(index='nama', columns='mobil')
fig, p = plt.subplots()
print(df)
plt.imshow(df)
plt.colorbar()

print(df.index)
col = list(map(lambda x: x[1], df.columns.tolist()))

index=list(map(lambda x: x, df.index.tolist()))
print(index)

p.set_xticks(np.arange(len(col)))
p.set_xticklabels(col)

p.set_yticks(np.arange(len(index)))
p.set_yticklabels(index)
plt.text(0,0,'makan')
plt.text(1,0,'makan')
plt.show()
# df[('jumlah','andi')]
# df[('jumlah', 'mobil a')]['andi']
```
