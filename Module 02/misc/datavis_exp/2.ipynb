```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,50)
dy=0.2
y=np.sin(x) + dy*np.random.randn(50)
plt.errorbar(x,y, yerr=dy, fmt='o')
```

```python
def f(x,y):
    return np.sin(x)**10 + np.cos(10+y*x)*np.cos(x)

x = np.linspace(0,5,50)
y = np.linspace(0,5,40)

X, Y = np.meshgrid(x,y)
# print(X, Y)
Z=f(X,Y)
plt.contourf(X,Y,Z,10, cmap='RdGy')
# plt.contourf()
plt.xlabel('ini X')
plt.ylabel('ini Y')
plt.title('ini Title')
# plt.cm.Rd
plt.colorbar()
```

```python
data=np.random.randn(100)
plt.hist(data, bins=20, alpha=0.5, color='steelblue', edgecolor='k', linestyle='-', histtype='stepfilled')
```

```python
x1=np.random.normal(0, 0.8,1000)
x2=np.random.normal(-2,1,1000)
x3=np.random.normal(3,2,1000)
kwargs={
    'histtype':'stepfilled',
    'alpha':0.3,
    'normed':True,
    'bins':40
}
plt.hist(x1, **kwargs)
# plt.plot(x1)
plt.hist(x2, **kwargs)
plt.hist(x3,**kwargs)
```
