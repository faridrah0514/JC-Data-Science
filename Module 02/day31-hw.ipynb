```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mysql.connector
plt.style.use('ggplot')
```

```python
con={
    'user': 'admin',
    'password': 'admin',
    'host': 'localhost',
    'database': 'world',
    'auth_plugin': 'mysql_native_password'
}
connector=mysql.connector.connect(**con)
fig=plt.figure(figsize=(17,12))
# print(connector)
df=pd.read_sql('select Name, Region, Population from country where region="southeast asia" order by name', connector)
df.set_index('Name', inplace=True)
plt.bar(df.index,df['Population'], color='tab:blue-orange-mediumseagreen-r-purple-sienna-orchid-grey-yellowgreen-darkturquoise'.split('-'), zorder=2,width=0.5)
plt.title("Populasi Negara ASEAN", fontsize=20)
plt.xlabel("Negara")
plt.ylabel("Populasi (x100jt jiwa)")
plt.xticks(rotation=45)
for i, j in zip((list(df.index)),(list(df['Population']))):
    plt.text(i, j*1.0188, f'{str(j)}',ha='center')
plt.grid(True)
plt.show()
```

```python
query="select name, region, population, (select sum(population) from country where region='southeast asia')  as total, ((population/((select sum(population) from country where region='southeast asia')))*100) as percentage from country where region='southeast asia' order by name"

df2=pd.read_sql(query, connector)
df2['percentage']
fig=plt.figure(figsize=(9,19))
_ , _ , autotexts=plt.pie(df2['percentage'], labels=df2['name'], autopct='%1.1f%%', shadow=True, colors='tab:blue-orange-mediumseagreen-r-purple-sienna-orchid-grey-yellowgreen-darkturquoise'.split('-')
)
for text in autotexts:
    text.set_color('white')
plt.title("Persentase Penduduk ASEAN", size=25)
plt.show()
```

```python
query = "select name, region, population, gnp from country where region='southeast asia' order by name;"
df3=pd.read_sql(query, connector)
fig=plt.figure(figsize=(14,10))
plt.bar(df3['name'], df3['gnp'], width=0.5, color='tab:blue-orange-mediumseagreen-r-purple-sienna-orchid-grey-yellowgreen-darkturquoise'.split('-'))
plt.title("Pendapatan Bruto Nasional ASEAN", fontsize=23)
plt.xlabel("Negara")
plt.ylabel("Gross National Product (US$)")
plt.xticks(rotation=45)
for i,j in zip(list(df3['name']),list(df3['gnp'])):
    plt.text(i,j+(10**3), str(j), ha='center')
plt.show()
```

```python
query="select name, surfacearea, (select sum(surfacearea) from country where region='southeast asia') as total, ((surfacearea/(select sum(surfacearea) from country where region='southeast asia'))*100) as percentage  from country where region='southeast asia' order by name"
df4=pd.read_sql(query, connector)
df4
fig=plt.figure(figsize=(9,19))
_ , _ , autotexts=plt.pie(df4['percentage'], labels=df4['name'], autopct='%1.1f%%', shadow=True, colors='tab:blue-orange-mediumseagreen-r-purple-sienna-orchid-grey-yellowgreen-darkturquoise'.split('-'))
for text in autotexts:
    text.set_color('white')
plt.title("Persentase Luas Daratan ASEAN", fontsize=23)
plt.show()
```

```python
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

mpl.rcParams['legend.fontsize'] = 10

fig = plt.figure()
ax = fig.gca(projection='3d')
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)
ax.plot(x, y, z, label='parametric curve')
ax.legend()

plt.show()

print(x)
print(y)
print(z)
```
