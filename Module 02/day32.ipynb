```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d
```

```python
p = plt.subplot(111, projection='3d')
x = np.arange(10)
y = np.arange(10)
z = np.ones(10)
dx= [2]*10
dy= [2]*10
dz = np.ones(10)

p.bar3d(x,y,z,dx,dy,dz)
p.set_xlabel('Sumbu X')
p.set_ylabel('sumbu Y')
p.set_zlabel('sumbu Z')
plt.show()
```
